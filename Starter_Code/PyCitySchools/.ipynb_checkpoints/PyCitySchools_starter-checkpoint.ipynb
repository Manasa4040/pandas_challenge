```python
# Dependencies and Setup
import pandas as pd
from pathlib import Path
```

```python
# File to Load 
school_data_to_load = Path("Resources/schools_complete.csv")
student_data_to_load = Path("Resources/students_complete.csv")
```

```python
# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)
```

```python
# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete
```

```python
#Calculate the percentage of students with a passing maths score (50 or greater):
perc_math_pass=(school_data_complete["maths_score"]>50).sum()/ school_data_complete["maths_score"].count()* 100
perc_math_pass 
```

```python
#Calculate the percentage of students with a passing reading score (50 or greater):
perc_reading_pass=(school_data_complete["reading_score"]>50).sum()/ school_data_complete["reading_score"].count()* 100
perc_reading_pass
```

# School Summary

* Create an overview table that summarises key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Maths Score
  * Average Reading Score
  * % Passing Maths
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed maths **and** reading.)
  
* Create a dataframe to hold the above results

```python
sch_count= len(school_data_complete["school_name"].unique())
sch_count
###school_count
```

```python
#School Name, lets set the schoolName & type col as index 1st
sch_types=school_data.set_index(["school_name"])["type"]
sch_types.head()
```

```python
#Total Students in each school
stu_per_sch=school_data_complete["school_name"].value_counts()
stu_per_sch.head()
```

```python
#Total School Budget
sch_budget=school_data_complete.groupby(["school_name"])["budget"].mean()
sch_budget.head()
```

```python
#Per Student Budget
per_stu_bud=sch_budget/stu_per_sch
per_stu_bud
```

```python
#Average Math Score per school
avg_math_perSch=school_data_complete.groupby(["school_name"])["maths_score"].mean()
avg_math_perSch.head()
```

```python
#Average Reading Score per school
avg_red_perSch= school_data_complete.groupby(["school_name"])["reading_score"].mean()
avg_red_perSch.head()
```

```python
#% Passing Math per school 
pas_math_scor=school_data_complete.loc[school_data_complete["maths_score"]>=50]
pas_math_scor

#assign the math scoor by each school
group_math_sch=pas_math_scor["school_name"].value_counts()
group_math_sch.head()

percent_math=group_math_sch/stu_per_sch*100
percent_math
```

```python
# % Passing Reading per school
read_scor=school_data_complete.loc[school_data_complete["reading_score"]>=50]

#assign the reading scoor by each school
by_read_PerSch= read_scor["school_name"].value_counts()
by_read_PerSch.head()

#get the final percentage result
percent_read=by_read_PerSch/stu_per_sch*100
percent_read
```

```python
# % Overall Passing Rate (Average of the above two)
overall=percent_math + percent_read/stu_per_sch
overall
```

```python
#creat a new DF for the School_summery

school_summery_DF = pd.DataFrame({"School Type":sch_types, "Total Students":stu_per_sch, "Total School Budget":sch_budget
                                ,"Per Student Budget":per_stu_bud ,"Average Math Score" :avg_math_perSch,
                             "Average Reading Score":avg_red_perSch,
                             "% Passing Math":percent_math, "% Passing Reading":percent_read,
                            "% Overall Passing Rate":overall})
school_summery_DF.head()
```

# Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
#Top Performing Schools (By Passing Rate)
TopSch_DF=school_summery_DF.sort_values(["% Passing Math","% Passing Reading","% Overall Passing Rate" ], 
                                        ascending=False)
TopSch_DF.head()
```

```python
#mapping Top Performing Schools

TopSch_DF["Total School Budget"]=TopSch_DF["Total School Budget"].map("${:,.2f}".format)
TopSch_DF["Per Student Budget"]=TopSch_DF["Per Student Budget"].map("${:,.2f}".format)


TopSch_DF.head()
```

# Bottom Performing Schools (By % Overall Passing)

Sort and display the five worst-performing schools by % overall passing.

```python
BottomSch_DF=school_summery_DF.sort_values("% Overall Passing Rate")
BottomSch_DF.head()
```

# Maths Scores by Year

Create a pandas series for each year. Hint: use a conditional statement.

Group each series by school

Combine the series into a dataframe

Optional: give the displayed data cleaner formatting

```python
# Function to create series for average math score by grade

def average_math_by_year(year):
    school_summary = school_data_complete.loc[school_data_complete.year == year].groupby(['school_name'])
    school_name = school_data_complete.school_name.sort_values().unique()
    school_avg_math_score = list(school_summary.maths_score.mean())

    # Compile all calculations into dataframe
    
    average_math_DF = pd.DataFrame({"School Names":school_name,
                                    f"{year} Average Math Score":school_avg_math_score})

    average_math_DF = average_math_DF.reset_index(drop=True)
    return average_math_DF

# Function to create series for average reading score by grade
def average_reading_by_year(year):
    school_summary = school_data_complete.loc[school_data_complete.year == year].groupby(['school_name'])
    school_name = school_data_complete.school_name.sort_values().unique()
    school_avg_reading_score = list(school_summary.reading_score.mean())

    # Compile all calculations into dataframe
    average_reading_DF = pd.DataFrame({"School Names":school_names,
                                    f"{year} Average Reading Score":school_avg_reading_score})

    average_reading_DF = average_reading_DF.reset_index(drop=True)
    return average_reading_DF
```
